```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("/content/drive/MyDrive/lok_sabha_data/Plot data/LS1_plot_data.csv")
```

```python
LOK_SABHA = [1 for i in range(len(df))]
print(LOK_SABHA)
```

```python
df['LOK_SABHA'] = LOK_SABHA
```

```python
df.head()
```

```python
for file in range(2, 18):
    temp_df = pd.read_csv("/content/drive/MyDrive/lok_sabha_data/Plot data/LS" + str(file) + "_plot_data.csv", header= 0, encoding= 'unicode_escape')
    LOK_SABHA = [file for l in range(len(temp_df))]
    temp_df['LOK_SABHA'] = LOK_SABHA
    df = pd.concat([df, temp_df])
```

```python
df.head()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
fig = px.scatter_geo(df, lat = "Latitude", lon = "Longitude", color="Frequency",
                     hover_name="Location", size="Frequency",
                     animation_frame="LOK_SABHA",
                     title = "Lok Sabha Data Analysis",
                     height = 700,
                     projection="natural earth")
fig.show()
```
